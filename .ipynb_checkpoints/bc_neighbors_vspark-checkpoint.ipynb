## Split the Data into Training and Testing Sets

### Step 1: Read the `breast-cancer.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
df = pd.read_csv(
    Path('Resources/breast-cancer.csv')   
)
# Review the DataFrame
df.head() 
```

```python
df['diagnosis'].astype('string')
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':0,'B':1})
df.tail()
```

```python
#List out the columns so you can make data frames using the the three classifications of data
df.columns
```

```python
#Creat three dataframes with each set of data. The Dataframes will be used to test the ability of the dataset to predict the diagnosis. 
#Therefore we, as developers, would be able to select the best data set to use for our model.
mean_df = df[['id', 'diagnosis', 'radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean']]

se_df = df[[ 'id', 'diagnosis','radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se']]

worst_df = df[['id', 'diagnosis','radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]

#Data Frame Check
mean_df.head()
```

```python
#Data Frame Check
se_df.head()
```

```python
#Data Frame Check
worst_df.head()
```

### Step 2: Create the labels set (`y`)  from the "diagnosis" column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features


# Separate the y variable, the labels. Used Loan Status as that is what the data is testing for.
y = worst_df['diagnosis']

# Separate the X variable, the features
X = worst_df.drop(columns='diagnosis')
```

### Step 3: Check the balance of the labels variable (`y`) by using the `value_counts` function.

```python
# Check the balance of our target values
y.value_counts()
```

### Step 4: Split the data into training and testing datasets by using `train_test_split`.

```python
from sklearn.preprocessing import StandardScaler
# Check the balance of our target values
X = pd.get_dummies(X)
X.head()

X_train, X_test, y_train, y_test = train_test_split(X, y)
# Instantiate a StandardScaler instance
scaler = StandardScaler()
# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)
# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)
# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Import the KNeighborsClassifier module from sklearn
from sklearn.neighbors import KNeighborsClassifier
# Instantiate the KNeighborsClassifier model with n_neighbors = 3
knn = KNeighborsClassifier(n_neighbors=5)
# Train the model using the training data
knn.fit(X_train_scaled, y_train)
# Create predictions using the testing data
y_pred = knn.predict(X_test_scaled)
# Print the classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred))
```

```python
# Create a method that creates a new Sequential model with hyperparameter options
def create_model(hp):
    nn_model = tf.keras.models.Sequential()
    # Allow kerastuner to decide which activation function to use in hidden layers
    activation = hp.Choice('activation',['relu','tanh','sigmoid'])
    # Allow kerastuner to decide number of neurons in first layer
    nn_model.add(tf.keras.layers.Dense(units=hp.Int('first_units',
        min_value=1,
        max_value=10,
        step=2), activation=activation, input_dim=11))
    # Allow kerastuner to decide number of hidden layers and neurons in hidden layers
    for i in range(hp.Int('num_layers', 1, 6)):
        nn_model.add(tf.keras.layers.Dense(units=hp.Int('units_' + str(i),
            min_value=1,
            max_value=10,
            step=2),
            activation=activation))
    nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
    # Compile the model
    nn_model.compile(loss="binary_crossentropy", optimizer='adam', metrics=["accuracy"])
    return nn_model
```

```python
# Import the kerastuner library
import keras_tuner as kt

tuner = kt.Hyperband(
    create_model,
    objective="val_accuracy",
    max_epochs=20,
    hyperband_iterations=2)
```

```python
# Run the kerastuner search for best hyperparameters
tuner.search(X_train_scaled,y_train,epochs=20,validation_data=(X_test_scaled,y_test))
```

```python
best_model = tuner.get_best_models(1)[0]
model_loss, model_accuracy = best_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
